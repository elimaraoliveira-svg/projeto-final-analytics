#9.Pandas na pr√°tica
O Pandas √© uma das bibliotecas mais populares do Python para an√°lise e manipula√ß√£o de dados.
Ele oferece duas estruturas principais:

Series: semelhantes a vetores ou colunas de uma tabela.

DataFrame: tabelas completas (linhas e colunas), muito parecidas com planilhas do Excel.

Com o Pandas √© poss√≠vel ler, transformar, filtrar, resumir e salvar dados de maneira simples e eficiente.

```python
# Instalando o Pandas (caso n√£o esteja no ambiente)
!pip install pandas

import pandas as pd

# -----------------------------
# 1. Criando um DataFrame simples
# -----------------------------
dados = {
    "Nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "Idade": [23, 35, 31, 29, 40],
    "Nota": [8.5, 7.0, 9.2, 6.8, 7.5]
}
df = pd.DataFrame(dados)

# -----------------------------
# 2. Visualizando os dados
# -----------------------------
print("Primeiras linhas do DataFrame:")
print(df.head())

print("\nColunas dispon√≠veis:")
print(df.columns)

print("\nEstat√≠sticas b√°sicas:")
print(df.describe())

# -----------------------------
# 3. Filtrando linhas (ex: alunos com nota maior que 7.5)
# -----------------------------
aprovados = df[df["Nota"] > 7.5]
print("\nAlunos com nota maior que 7.5:")
print(aprovados)

# -----------------------------
# 4. Adicionando uma nova coluna derivada
# -----------------------------
df["Aprovado"] = df["Nota"] >= 7
print("\nDataFrame com coluna 'Aprovado':")
print(df)

# -----------------------------
# 5. Salvando o DataFrame em CSV
# -----------------------------
df.to_csv("alunos.csv", index=False)
print("\nArquivo 'alunos.csv' salvo com sucesso!")
```

```python
# Instalando o Pandas
!pip install pandas

import pandas as pd

# -----------------------------
# 1. Criando um CSV fict√≠cio (simula√ß√£o de base de vendas)
# -----------------------------
dados_vendas = {
    "Produto": ["Notebook", "Notebook", "Smartphone", "Smartphone", "Tablet", "Tablet", "Headphone", "Headphone"],
    "Quantidade": [2, 1, 3, 5, 2, 4, 6, 2],
    "Pre√ßo_Unitario": [3500, 3500, 1200, 1200, 1500, 1500, 200, 200],
    "Vendedor": ["Ana", "Bruno", "Carlos", "Ana", "Daniela", "Eduardo", "Carlos", "Ana"]
}

df_vendas = pd.DataFrame(dados_vendas)
df_vendas.to_csv("vendas.csv", index=False)  # salvando CSV para simular leitura

# -----------------------------
# 2. Lendo o CSV
# -----------------------------
df = pd.read_csv("vendas.csv")
print("Base de dados carregada:")
print(df.head())

# -----------------------------
# 3. Criando uma nova coluna (Valor Total = Quantidade * Pre√ßo_Unitario)
# -----------------------------
df["Valor_Total"] = df["Quantidade"] * df["Pre√ßo_Unitario"]

print("\nDataFrame com coluna 'Valor_Total':")
print(df)

# -----------------------------
# 4. Estat√≠sticas e agrega√ß√µes
# -----------------------------
print("\nResumo estat√≠stico:")
print(df.describe())

# Total de vendas por produto
vendas_produto = df.groupby("Produto")["Valor_Total"].sum()
print("\nTotal de vendas por produto:")
print(vendas_produto)

# M√©dia de vendas por vendedor
media_vendas_vendedor = df.groupby("Vendedor")["Valor_Total"].mean()
print("\nM√©dia de vendas por vendedor:")
print(media_vendas_vendedor)

# -----------------------------
# 5. Exportando os resultados
# -----------------------------
vendas_produto.to_csv("total_vendas_produto.csv")
media_vendas_vendedor.to_csv("media_vendas_vendedor.csv")

print("\nArquivos CSV exportados com sucesso!")
```

#10.Estat√≠stica (descri√ß√µes + gr√°ficos)

A estat√≠stica descritiva √© usada para resumir, organizar e interpretar dados de forma objetiva.
Ela ajuda a identificar tend√™ncias, padr√µes e varia√ß√µes nos dados atrav√©s de medidas como:

M√©dia: valor m√©dio dos dados.

Mediana: valor central que divide os dados em duas partes iguais.

Desvio padr√£o: mede a dispers√£o dos dados em rela√ß√£o √† m√©dia.

Quartis: valores que dividem os dados em quatro partes iguais.

Al√©m das medidas num√©ricas, gr√°ficos s√£o essenciais para visualizar os dados:

Histograma: mostra a distribui√ß√£o de uma vari√°vel num√©rica.

Gr√°fico de barras: compara valores de diferentes categorias.

Gr√°fico de linhas: evidencia tend√™ncias ao longo de uma sequ√™ncia ou tempo.

Boxplot: visualiza dispers√£o, mediana e valores at√≠picos.

Com Python, usamos Pandas para manipula√ß√£o de dados e Matplotlib/Seaborn para visualiza√ß√£o.

```python
# Instalando bibliotecas (caso necess√°rio)
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Criando um DataFrame fict√≠cio
# -----------------------------
dados = {
    "Idade": [23, 35, 31, 29, 40, 28, 33, 37, 30, 25],
    "Salario": [3500, 4200, 3900, 3100, 5000, 3300, 4100, 4500, 3800, 3600],
    "Departamento": ["Vendas", "RH", "TI", "TI", "Vendas", "RH", "TI", "Vendas", "RH", "TI"]
}

df = pd.DataFrame(dados)

# -----------------------------
# 2. Estat√≠sticas descritivas
# -----------------------------
print("Resumo estat√≠stico:\n", df.describe())
print("\nM√©dia de idade:", df["Idade"].mean())
print("Mediana do sal√°rio:", df["Salario"].median())
print("Desvio padr√£o do sal√°rio:", df["Salario"].std())

# -----------------------------
# 3. Gr√°ficos
# -----------------------------

# Histograma da Idade
plt.figure(figsize=(6,4))
plt.hist(df["Idade"], bins=5, color="skyblue", edgecolor="black")
plt.title("Distribui√ß√£o de Idade")
plt.xlabel("Idade")
plt.ylabel("Frequ√™ncia")
plt.show()

# Gr√°fico de barras - m√©dia salarial por Departamento
plt.figure(figsize=(6,4))
media_salario = df.groupby("Departamento")["Salario"].mean()
media_salario.plot(kind="bar", color="salmon")
plt.title("M√©dia Salarial por Departamento")
plt.ylabel("Sal√°rio M√©dio")
plt.show()

# Boxplot do Sal√°rio
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Salario"], color="lightgreen")
plt.title("Boxplot do Sal√°rio")
plt.show()
```

#11.An√°lise Gr√°fica
A an√°lise gr√°fica ou visualiza√ß√£o de dados √© uma ferramenta essencial para compreender informa√ß√µes de forma r√°pida e intuitiva.
Ela permite identificar padr√µes, tend√™ncias, distribui√ß√µes e outliers que muitas vezes n√£o aparecem em tabelas ou estat√≠sticas simples.

Os tipos de gr√°ficos mais usados incluem:

Gr√°fico de barras: compara valores entre categorias.

Gr√°fico de linhas: mostra evolu√ß√£o ou tend√™ncia ao longo do tempo.

Gr√°fico de dispers√£o (scatter plot): analisa rela√ß√£o entre duas vari√°veis num√©ricas.

Gr√°fico de pizza (pie chart): mostra propor√ß√£o de partes em rela√ß√£o ao todo.

Boxplot: visualiza dispers√£o, mediana e valores at√≠picos.

O estudo de caso analisa a realidade de uma ag√™ncia digital do Sicoob, focada em vendas via atendimento remoto.
O principal desafio √© conseguir efetivar vendas sem contato presencial, mantendo a satisfa√ß√£o do cooperado e garantindo resultados.

Desafios identificados:

```python
# Instalando bibliotecas (caso necess√°rio)
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Criando DataFrame fict√≠cio
# -----------------------------
dados = {
    "Cliente": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo", "Fernanda", "Gustavo", "Helena", "Igor", "Joana"],
    "Idade": [23, 35, 31, 29, 40, 28, 33, 37, 30, 25],
    "Canal": ["WhatsApp", "App", "Chat", "Call", "WhatsApp", "App", "Chat", "Call", "WhatsApp", "App"],
    "Produto": ["Cr√©dito", "Seguro", "Investimento", "Cr√©dito", "Investimento", "Cr√©dito", "Seguro", "Investimento", "Seguro", "Cr√©dito"],
    "Valor_Venda": [5000, 1200, 15000, 7000, 10000, 6000, 1300, 16000, 1400, 8000],
    "Status": ["Conclu√≠da", "Conclu√≠da", "Conclu√≠da", "Pendente", "Conclu√≠da", "Conclu√≠da", "Cancelada", "Conclu√≠da", "Pendente", "Conclu√≠da"]
}

df = pd.DataFrame(dados)

# -----------------------------
# 2. An√°lise b√°sica
# -----------------------------
# Quantidade de atendimentos por canal
atendimentos_canal = df["Canal"].value_counts()
print("Atendimentos por canal:\n", atendimentos_canal)

# Taxa de convers√£o por canal
concluidas_canal = df[df["Status"]=="Conclu√≠da"].groupby("Canal").size()
total_canal = df.groupby("Canal").size()
taxa_conversao = (concluidas_canal / total_canal * 100).round(2)
print("\nTaxa de convers√£o por canal (%):\n", taxa_conversao)

# Total de vendas por produto
vendas_produto = df.groupby("Produto")["Valor_Venda"].sum()
print("\nTotal de vendas por produto:\n", vendas_produto)

# M√©dia de vendas por faixa et√°ria (ex: <30 e >=30)
df["Faixa_Idade"] = df["Idade"].apply(lambda x: "<30" if x<30 else ">=30")
media_venda_faixa = df.groupby("Faixa_Idade")["Valor_Venda"].mean()
print("\nM√©dia de vendas por faixa et√°ria:\n", media_venda_faixa)

# -----------------------------
# 3. Gr√°ficos
# -----------------------------
# Gr√°fico de barras por canal
plt.figure(figsize=(6,4))
sns.countplot(x="Canal", data=df, palette="pastel")
plt.title("Quantidade de Atendimentos por Canal")
plt.show()

# Gr√°fico de pizza por produto
plt.figure(figsize=(6,6))
plt.pie(vendas_produto, labels=vendas_produto.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Total de Vendas por Produto")
plt.show()

# Histograma de valores de venda
plt.figure(figsize=(6,4))
plt.hist(df["Valor_Venda"], bins=5, color="lightblue", edgecolor="black")
plt.title("Distribui√ß√£o dos Valores de Venda")
plt.xlabel("Valor da Venda")
plt.ylabel("Frequ√™ncia")
plt.show()
```
